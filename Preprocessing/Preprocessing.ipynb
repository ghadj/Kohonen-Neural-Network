```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
SRC_PATH = "./letter-recognition.txt"
DEST_PATH = "./"
```

```python
def normalize(df):
    result = df.copy()
    for c in df.columns[1:]:
        max_value = df[c].max()
        min_value = df[c].min()
        result[c] = (df[c] - min_value) / (max_value - min_value)
    return result
```

```python
def generateTarget(df):
    df_target = pd.DataFrame()
    for index, row in df.iterrows():
        zeroes = list([0]*26)
        zeroes[ord(row[0])- 65] = 1
        df_target = df_target.append([zeroes])
    return df_target.reset_index(drop=True)
```

```python
df_training = pd.DataFrame()
df_test = pd.DataFrame()
```

```python
df = pd.read_csv(SRC_PATH,header=None)
df
```

```python
df_norm = normalize(df)
df_norm
```

```python
for letter in range(65,91):
    df_selected = df_norm.loc[df_norm[0] == chr(letter)]
    
    # Randomly sample 70% of dataframe containing only the selected letter
    df_percent = df_selected.sample(frac=0.7)
    df_rest = df_selected.loc[~df_selected.index.isin(df_percent.index)]
    
    df_training = df_training.append(df_percent, ignore_index = True)
    df_test = df_test.append(df_rest, ignore_index = True)
```

```python
df_training.reset_index(drop=True)
df_test.reset_index(drop=True)

#df_training = pd.concat([df_training, generateTarget(df_training)], axis=1)
#df_test = pd.concat([df_test, generateTarget(df_test)], axis=1)

df_test.sample(frac=1).to_csv(DEST_PATH + "test.csv", index=False, header=None)
df_training.sample(frac=1).to_csv(DEST_PATH + "training.csv", index=False, header=None)
```
